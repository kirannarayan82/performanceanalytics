```python
deliveriesdata=pd.read_csv('/home/kiran/Documents/ipl/deliveries.csv')
```

```python
#Just like any data , this data needs to be cleaned up.lets clean the data and treat them for missing values
```

```python
deliveriesdata['player_dismissed']
```

```python
deliveriesdata=deliveriesdata.fillna(0)#All missing vlaues are treated to 0
```

```python
dismissals=deliveriesdata[deliveriesdata['batsman']==deliveriesdata['player_dismissed']]
```

```python
dismissals
```

```python
dismissals_bowler=dismissals.groupby(['bowler']).size().reset_index()
```

```python
#We will first see the number of wickets taken by each bowler
```

```python
dismissals_bowler
```

```python
#We will see the number of balls bowled by each bowler
```

```python
dismissals_ballsbowled=deliveriesdata.groupby(['bowler'])['ball'].count().reset_index()
```

```python
dismissals_ballsbowled
```

```python
#Next we will see the runs given by each bowler
```

```python
dismissals_runsgiven=deliveriesdata.groupby(['bowler'])['total_runs'].sum().reset_index()
```

```python
dismissialsbolwerruns=pd.merge(dismissals_ballsbowled,dismissals_runsgiven)
```

```python
#Combining the runs given and the wickets taken per bowler
```

```python
wicketsperbowler=pd.merge(dismissialsbolwerruns,dismissals_bowler)
```

```python
wicketsperbowler
```

```python
wicketsperbowler.columns
```

```python
wicketsperbowler.columns=['bowler','ballsbowled','totalrunsconceded','numberofdismissals']
```

```python
wicketsperbowler
```

```python
wicketsperbowler.sort_values('numberofdismissals',ascending=False)
```

Lasith Malinga is the highest wickettaker in IPL. He has also conceded the least runs per wicket. That means he is a real asset to any team as a bowler. No wonder he is so highly rated in IPL

```python
#Lets look at the batsman , the runs they scored ,the number of balls they faced
```

```python
batsmandata=deliveriesdata.groupby(['batsman'])['total_runs'].sum().reset_index()
```

```python
batsmandata_ballsfaced=deliveriesdata.groupby(['batsman'])['ball'].count().reset_index()
```

```python
batsmandata_ballsfaced
```

```python
batsmandata=pd.merge(batsmandata,batsmandata_ballsfaced,left_on='batsman',right_on='batsman',how='inner')
```

```python
batsmandata=pd.merge(batsmandata,dismissals,left_on='batsman',right_on='batsman',how='inner')
```

```python
batsmandata.columns
```

```python
batsmandata.columns=['batsman', 'total_runs','numberofballsfaced','numberofdismissals']
```

```python
#Looking at all batsman ,their runs scored ,the number of balls they faced and the number of times they have got out
```

```python
batsmandata
```

```python
batsmandata['strikerate']=(batsmandata['total_runs']/batsmandata['numberofballsfaced'])*100
```

```python
batsmandata['average']=batsmandata['total_runs']/batsmandata['numberofdismissals']
```

```python
batsmandata['strikerate'].describe()
```

```python
batsmandata=batsmandata[batsmandata['numberofballsfaced']>100]
```

```python
topstrikerates=batsmandata.sort_values(by='strikerate',ascending=False)
```

```python
#We will get the strike rate and average of each batsman. The strike rate is an important parameter in a fast paced game like IPL. The quicker you score , the greater the value a batsman adds to the team
```

```python
topstrikerates.head(20)
```

We see that Andre Russel has the highest strike rate across all matches,considering all batsman who have faced more than 100 balls

```python
average=batsmandata.sort_values(by='average',ascending=False)
```
